```python
# Librerías para manejo de datos
import pandas as pd
import numpy as np
```

## Carga de Datos

En esta unidad, se presentan el perfilamiento y preparación de los datos para la ejecución de los modelos. 

### Dimensión ciudad


```python
# Se cargan los datos. 
dfCity = pd.read_csv('dimension_city.csv', sep=',')
dfCity
```

```python
# En esta dimensión se va a eliminar la primera fila (unknown), ya que 
# no está referenciada desde la tabla de hechos. Además, se elimina la 
# columna rowID, teniendo en cuenta que no aporta información. 
# A pesar de que la población no tiene datos completos, no se puede obviar

dfCity.loc[dfCity.City_Key == 0, "City_Key"] = np.nan
dfCity.dropna(inplace = True)
dfCity.drop(['row ID'], axis=1, inplace=True)
dfCity
```

### Dimensión Customer




```python
# Se cargan los datos
dfCustomer = pd.read_csv('dimension_customer.csv', sep=',', encoding='latin-1')
dfCustomer
```

```python
# Se elimina la primera fila (unknown), dado que no está siendo llamada
# desde la tabla de hechos
dfCustomer.loc[dfCustomer.Customer_Key == 0, "Customer_Key"] = np.nan
dfCustomer.dropna(inplace = True)
dfCustomer
```

### Dimensión Date

```python
# Se cargan los datos
# Al hacer la inspección, se puede ver que no se necesita modificación alguna
dfDate = pd.read_csv('dimension_date.csv', sep=',', encoding='latin-1')
dfDate
```

### Dimensión Employee

```python
# Se cargan los datos
dfEmployee = pd.read_csv('dimension_employee.csv', sep=',', encoding='latin-1')
dfEmployee
```

```python
# La limpieza de los datos en esta parte se asocia a eliminar la primera fila
# y cambiar el atributo binario por valores 1, 0
dfEmployee.loc[dfEmployee.Employee_Key == 0, "Employee_Key"] = np.nan
dfEmployee.dropna(inplace = True)
dfEmployee
```

### Dimensión Stock Item

```python
# Se cargan los datos
dfStockItem = pd.read_csv('dimension_stock_item.csv', sep=',', encoding='latin-1')
dfStockItem.head(5)
```

```python
# Se elimina la primera fila dado que es el desconocido y nunca está 
# referenciada dentro de la tabla de hechos
dfStockItem = dfStockItem.drop(dfStockItem.index[0])
dfStockItem.head(5)
```

```python
# Se revisan los valores de la variable color:
# Se puede ver que hay muchos datos nulos,no obstante, puede ser valioso 
# realizar alguna clasificación dentro de la tabla de hechos para los demás 
# colores, así que serán conservados estos valores
dfStockItem['Color'].value_counts()
```

```python
# Se revisa lo mismo para selling package, pero nuevamente se puede ver que es 
# importante que se puedan hacer análisis sobre todos los datos:
dfStockItem['Selling_Package'].value_counts()
```

```python
# Viendo la columna Brand, se puede ver que tiene un gran porcentaje de nulidad, 
# y los valores que no son nulos solamente tienen un valor, por lo cual se 
# considera que esto no aporta valor y por tanto debe ser eliminado
dfStockItem['Brand'].value_counts()
```

```python
# Posteriormente, se analiza la columna size_val, en donde podemos ver que 
# a pesar de los nulos, hay información valiosa, por lo cual no se hace ninguna
# modificación
dfStockItem['Size_val'].value_counts()
```

```python
# Después se revisa el lead time , pero no se hace ninguna modificación
dfStockItem['Lead_Time_Days'].value_counts()
```

```python
# Posteriormente, se revisa Quantity_Per_Outer. 
# No hay faltantes, por lo cual no se modifica la variable
dfStockItem['Quantity_Per_Outer'].value_counts()
```

```python
# Ahora, se revisa el Tax_Rate, que debe ser numérico:
dfStockItem['Tax_Rate'].describe()
```

```python
# Se deben reemplazar las comas por puntos para así tener una representación
# numérica adecuada
aux = []
for valor in dfStockItem['Tax_Rate']:
  aux.append( valor.replace(',','.'))

dfStockItem['Tax_Rate'] = aux
dfStockItem['Tax_Rate'] = dfStockItem['Tax_Rate'].astype('float')
dfStockItem['Tax_Rate'].describe()
```

```python
# Se debe hacer lo mismo para la columna de Unit_Price:
dfStockItem['Unit_Price'].describe()
```

```python
# Se deben reemplazar las comas por puntos para así tener una representación
# numérica adecuada
aux = []
for valor in dfStockItem['Unit_Price']:
  aux.append( valor.replace(',','.'))

dfStockItem['Unit_Price'] = aux
dfStockItem['Unit_Price'] = dfStockItem['Unit_Price'].astype('float')
dfStockItem['Unit_Price'].describe()
```

```python
# Se debe hacer lo mismo para el recommended retail price
dfStockItem['Recommended_Retail_Price'].describe()
```

```python
# Se deben reemplazar las comas por puntos para así tener una representación
# numérica adecuada
aux = []
for valor in dfStockItem['Recommended_Retail_Price']:
  aux.append(valor.replace(',','.'))

dfStockItem['Recommended_Retail_Price'] = aux 
dfStockItem['Recommended_Retail_Price'] = dfStockItem['Recommended_Retail_Price'].astype('float')
dfStockItem['Recommended_Retail_Price'].describe()
```

```python
# Finalmente, se hace para Typical_Weight_Per_Unit
dfStockItem['Typical_Weight_Per_Unit'].describe()
```

```python
# Se deben reemplazar las comas por puntos para así tener una representación
# numérica adecuada
aux = []
for valor in dfStockItem['Typical_Weight_Per_Unit']:
  aux.append(valor.replace(',','.'))

dfStockItem['Typical_Weight_Per_Unit'] = aux 
dfStockItem['Typical_Weight_Per_Unit'] = dfStockItem['Typical_Weight_Per_Unit'].astype('float')
dfStockItem['Typical_Weight_Per_Unit'].describe()
```
